```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Upload the dataset
# Replace 'your_dataset.csv' with the path to your dataset file
df = pd.read_csv('/Users/ananyasmaranikadivendi/Downloads/Baltimore911.csv')

# Step 2: Calculate non-null and missing values per column
non_null_counts = df.notnull().sum()
missing_counts = df.isnull().sum()

# Step 3: Prepare data for plotting
summary = pd.DataFrame({
    'Non-Null Values': non_null_counts,
    'Missing Values': missing_counts
})

# Step 4: Plot the data
summary.plot(kind='bar', figsize=(14, 7))

# Add title and labels
plt.title('Non-Null and Missing Values per Column')
plt.xlabel('Columns')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(loc='upper right')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Upload the dataset
# Replace 'your_dataset.csv' with the path to your dataset file
df = pd.read_csv('/Users/ananyasmaranikadivendi/Downloads/Baltimore911.csv')

# Step 2: Identify column types
column_types = df.dtypes

# Classify column types
type_counts = column_types.value_counts()

# Define a list of columns and their categories manually
# For demonstration purposes, adjust according to your dataset
temporal_columns = ['CrimeDate', 'CrimeTime']  # Example temporal columns
spatial_columns = ['Longitude', 'Latitude', 'Location']  # Example spatial columns

# Determine the type of each column
categories = []
for col in df.columns:
    if col in temporal_columns:
        categories.append('Temporal')
    elif col in spatial_columns:
        categories.append('Spatial')
    elif df[col].dtype == 'object':
        categories.append('Symbolic/Discrete')
    else:
        categories.append('Numeric/Real/Continuous')

# Create a DataFrame for category counts
category_counts = pd.Series(categories).value_counts()

# Step 3: Plot column types
plt.figure(figsize=(12, 6))

# Plot for column types
plt.subplot(1, 2, 1)
type_counts.plot(kind='bar', color='skyblue')
plt.title('Column Types')
plt.xlabel('Type')
plt.ylabel('Count')

# Plot for column categories
plt.subplot(1, 2, 2)
category_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue', 'lightgreen'])
plt.title('Column Categories')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Upload the dataset
# Replace 'your_dataset.csv' with the path to your dataset file
df = pd.read_csv('/Users/ananyasmaranikadivendi/Downloads/Baltimore911.csv')

# Step 2: Identify discrete attributes
discrete_attrs = [col for col in df.columns if pd.api.types.is_object_dtype(df[col])]

# Step 3: Analyze discrete attributes
unique_counts = {}
top_values = {}

for attr in discrete_attrs:
    unique_counts[attr] = df[attr].nunique()  # Count unique values
    top_values[attr] = df[attr].value_counts().head(3)  # Top 3 most common values

# Step 4: Create a DataFrame for plotting unique value counts
unique_counts_df = pd.DataFrame(list(unique_counts.items()), columns=['Attribute', 'Unique Values'])

# Step 5: Plot unique values per attribute
plt.figure(figsize=(12, 6))
sns.barplot(x='Attribute', y='Unique Values', data=unique_counts_df, palette='viridis')
plt.title('Number of Unique Values for Discrete Attributes')
plt.xlabel('Discrete Attributes')
plt.ylabel('Number of Unique Values')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Step 6: Display top 3 values for each discrete attribute
for attr, values in top_values.items():
    print(f"\nAttribute: {attr}")
    print(values)
```

```python


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Upload the dataset
# Replace 'your_dataset.csv' with the path to your dataset file
df = pd.read_csv('/Users/ananyasmaranikadivendi/Downloads/Baltimore911.csv')

# Step 2: Calculate the percentage of rows with missing values per column
missing_percentages = df.isnull().mean() * 100  # Calculate the percentage of missing values

# Step 3: Create a DataFrame for plotting
missing_df = pd.DataFrame({
    'Column': missing_percentages.index,
    'Missing Percentage': missing_percentages.values
})

# Step 4: Plot the missing values percentage
plt.figure(figsize=(12, 6))
sns.barplot(x='Column', y='Missing Percentage', data=missing_df, palette='viridis')
plt.title('Percentage of Rows with Missing Values per Column')
plt.xlabel('Columns')
plt.ylabel('Percentage of Missing Values')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
