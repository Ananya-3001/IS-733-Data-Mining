# Column Types and categories

```python
import pandas as pd

# Load your dataset
# Adjust the path as needed
df = pd.read_csv('/Users/ananyasmaranikadivendi/Downloads/Baltimore911.csv')

# Define a function to categorize columns
def categorize_column(column):
    if pd.api.types.is_numeric_dtype(column):
        return 'Numeric/Real/Continuous'
    elif pd.api.types.is_datetime64_any_dtype(column):
        return 'Temporal'
    elif pd.api.types.is_string_dtype(column):
        return 'Symbolic/Discrete'
    else:
        return 'Other'

# Create a DataFrame to store the column categorization
column_types = pd.DataFrame({
    'Column': df.columns,
    'Type': [categorize_column(df[col]) for col in df.columns]
})

print(column_types)
```

# Summary of Numeric Attributes: Min, Max, Mean, Median, and Standard Deviation

```python
import pandas as pd

# Load your dataset
# Adjust the path as needed
df = pd.read_csv('/Users/ananyasmaranikadivendi/Downloads/Baltimore911.csv')

# Create a DataFrame to store statistics for numeric columns
numeric_stats = pd.DataFrame()

# Loop through columns and calculate statistics if the column is numeric
for col in df.columns:
    if pd.api.types.is_numeric_dtype(df[col]):
        numeric_stats[col] = pd.Series({
            'Min': df[col].min(),
            'Max': df[col].max(),
            'Mean': df[col].mean(),
            'Median': df[col].median(),
            'Std Dev': df[col].std()
        })

print(numeric_stats)
```

# Summary of Discrete Attributes: Unique Values and Most Frequent Values

```python
import pandas as pd

# Load your dataset
# Adjust the path as needed
df = pd.read_csv('/Users/ananyasmaranikadivendi/Downloads/Baltimore911.csv')

# Create a DataFrame to store statistics for discrete columns
discrete_stats = pd.DataFrame()

# Loop through columns and calculate statistics if the column is discrete
for col in df.columns:
    if pd.api.types.is_string_dtype(df[col]) or pd.api.types.is_categorical_dtype(df[col]):
        unique_count = df[col].nunique()
        top_three = df[col].value_counts().head(3)
        discrete_stats[col] = pd.Series({
            'Unique Count': unique_count,
            'Top 1 Value': top_three.index[0] if len(top_three) > 0 else None,
            'Top 1 Count': top_three.values[0] if len(top_three) > 0 else None,
            'Top 2 Value': top_three.index[1] if len(top_three) > 1 else None,
            'Top 2 Count': top_three.values[1] if len(top_three) > 1 else None,
            'Top 3 Value': top_three.index[2] if len(top_three) > 2 else None,
            'Top 3 Count': top_three.values[2] if len(top_three) > 2 else None,
        })

print(discrete_stats)
```

# Task 2: 

# series of plots to describe the temporal pattern (year-to-year, monthly, and day-of-week) of the overall crime incidence aggregating from all geo-locations

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/Users/ananyasmaranikadivendi/Downloads/Baltimore911.csv')

# Combine 'CrimeDate' and 'CrimeTime' into a single datetime column
df['CrimeTime'] = df['CrimeTime'].fillna('00:00:00')  # Default time if missing
df['DateTime'] = pd.to_datetime(df['CrimeDate'].astype(str) + ' ' + df['CrimeTime'])

# 1. Yearly Crime Counts
df['year'] = df['DateTime'].dt.year
yearly_counts = df.groupby('year').size()

# Plot yearly crime counts
plt.figure(figsize=(12, 6))
yearly_counts.plot(kind='bar', width=0.8)
plt.title('Yearly Crime Incidence')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# 2. Monthly Crime Counts
# Extract year and month, then aggregate crime counts by month
df['year_month'] = df['DateTime'].dt.to_period('M')
monthly_counts = df.groupby('year_month').size()

# Convert the 'year_month' period to string for clearer x-axis labels
monthly_counts.index = monthly_counts.index.astype(str)

# Plot monthly crime counts with formatted x-axis labels
plt.figure(figsize=(12, 6))
monthly_counts.plot(kind='bar', width=0.8)
plt.title('Monthly Crime Incidence')
plt.xlabel('Year-Month')
plt.ylabel('Number of Crimes')

# Adjust the number of x-ticks shown and rotate labels
plt.xticks(ticks=range(0, len(monthly_counts), max(len(monthly_counts)//12, 1)), 
           labels=monthly_counts.index[::max(len(monthly_counts)//12, 1)], rotation=45)

plt.grid(axis='y')
plt.tight_layout()
plt.show()

# 3. Day of the Week Crime Counts
# Extract day of the week and aggregate crime counts by day
df['day_of_week'] = df['DateTime'].dt.day_name()
weekly_counts = df.groupby('day_of_week').size()

# Reorder days of the week for better visualization
order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekly_counts = weekly_counts.reindex(order)

# Plot day-of-the-week crime counts
plt.figure(figsize=(12, 6))
weekly_counts.plot(kind='bar', width=0.8)
plt.title('Crime Incidence by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# Task 4

# A series of plots to illustrate how crime type distributions might vary from year-to-year, month-to-month, or by day-of-week aggregating across all locations.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/Users/ananyasmaranikadivendi/Downloads/Baltimore911.csv')

# Convert 'CrimeDate' and 'CrimeTime' to datetime format
df['DateTime'] = pd.to_datetime(df['CrimeDate'] + ' ' + df['CrimeTime'], format='%m/%d/%Y %H:%M:%S')

# Extract year, month, and day of the week from the DateTime column
df['Year'] = df['DateTime'].dt.year
df['Month'] = df['DateTime'].dt.month
df['DayOfWeek'] = df['DateTime'].dt.day_name()

# Function to plot crime type distribution by a given time period
def plot_crime_distribution_by_time_period(df, period, period_name):
    plt.figure(figsize=(12, 8))
    crime_distribution = df.groupby([period, 'Description']).size().unstack().fillna(0)
    crime_distribution.plot(kind='bar', stacked=True, colormap='tab20', figsize=(12, 8))
    plt.title(f'Crime Type Distribution by {period_name}')
    plt.xlabel(period_name)
    plt.ylabel('Number of Incidents')
    plt.legend(title='Crime Type', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xticks(rotation=90)
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()

# Plot distribution by year
plot_crime_distribution_by_time_period(df, 'Year', 'Year')

# Plot distribution by month
plot_crime_distribution_by_time_period(df, 'Month', 'Month')

# Plot distribution by day of the week
plot_crime_distribution_by_time_period(df, 'DayOfWeek', 'Day of the Week')
```

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Load your dataset
df = pd.read_csv('/Users/ananyasmaranikadivendi/Downloads/Baltimore911.csv')
df['CrimeDate'] = pd.to_datetime(df['CrimeDate'], errors='coerce')
df['Year'] = df['CrimeDate'].dt.year
df['Month'] = df['CrimeDate'].dt.month
df['DayOfWeek'] = df['CrimeDate'].dt.day_name()

# Initialize the Dash app
app = dash.Dash(__name__)

# App layout
app.layout = html.Div([
    html.H1("Crime Data Dashboard"),
    
    dcc.Dropdown(
        id='time-period-dropdown',
        options=[
            {'label': 'Year', 'value': 'Year'},
            {'label': 'Month', 'value': 'Month'},
            {'label': 'Day of Week', 'value': 'DayOfWeek'}
        ],
        value='Year'
    ),
    
    dcc.Dropdown(
        id='location-dropdown',
        options=[{'label': loc, 'value': loc} for loc in df['Location'].unique()],
        value=df['Location'].unique()[0]
    ),
    
    dcc.Graph(id='crime-type-bar-plot'),
    dcc.Graph(id='crime-map')
])

# Callbacks to update the dashboard based on user input
@app.callback(
    [Output('crime-type-bar-plot', 'figure'),
     Output('crime-map', 'figure')],
    [Input('time-period-dropdown', 'value'),
     Input('location-dropdown', 'value')]
)
def update_dashboard(time_period, location):
    # Filter data based on location
    filtered_df = df[df['Location'] == location]
    
    # Group data by selected time period
    time_period_data = filtered_df.groupby(time_period)['Total Incidents'].sum().sort_values(ascending=False)
    
    # Create crime type distribution bar plot
    crime_type_fig = px.bar(
        time_period_data,
        x=time_period_data.index,
        y=time_period_data.values,
        labels={'x': time_period, 'y': 'Total Incidents'},
        title=f'Crime Type Distribution by {time_period}'
    )
    
    # Create a map plot
    crime_map_fig = go.Figure(go.Scattermapbox(
        lat=filtered_df['Latitude'],
        lon=filtered_df['Longitude'],
        mode='markers',
        marker=dict(size=5, color='red'),
        text=filtered_df['Description']
    ))
    
    crime_map_fig.update_layout(
        mapbox_style="open-street-map",
        margin={"r":0,"t":0,"l":0,"b":0},
        title='Crime Incidents Map'
    )
    
    return crime_type_fig, crime_map_fig

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
